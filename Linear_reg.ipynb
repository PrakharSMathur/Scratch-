```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
df=pd.read_excel('Data/fire_theft.xls')
X_train,Y_train=np.array(df['X']),np.array(df['Y'])
plt.scatter(X_train,Y_train)
```

```python
def forward(W,X,b):
    Y=W*X+b
    return Y
```

```python
W,b=0,1
cost=0
```

```python
def optimize(X_train,Y_train,W,al,b,cost):
    l=[]
    for i in range(0,len(X_train)):
        new_W=W+al*(Y_train[i]-forward(W,X_train[i],b))
        W=new_W
        loss=0.5*(((forward(W,X_train[i],b))-Y_train[i])**2)
        l.append(loss)
        cost=cost+loss
    return new_W,l
```

```python
W,l=optimize(X_train,Y_train,0,0.01,1,cost)
```

```python
#plot loss
plt.plot(l[0:len(l)], 'ro-')
```

```python
Y_hat=[]
for i in X_train:
    Y_hat.append(W*i+b)
```

```python
plt.plot(X_train[0:len(X_train)], Y_hat[0:len(Y_hat)], 'ro-')
plt.scatter(X_train,Y_train)
```
